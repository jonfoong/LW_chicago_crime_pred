## Spatial Data

```python
import geopandas as gpd
from shapely.geometry import Point


#https://data.cityofchicago.org/Facilities-Geographic-Boundaries/Boundaries-Community-Areas-current-/cauq-8yn6
```

```python
#Load the geoJSON file
communities = gpd.read_file('Boundaries - Community Areas.geojson')
communities
```

```python
query_spatial = """
SELECT Date, ID, CAST(Longitude AS FLOAT64) as Longitude, CAST(Latitude AS FLOAT64) as Latitude,`Community Area`
FROM `wagon-bootcamp-428814.chicago_crime.chicago_crime_tab`
WHERE `Community Area` IS NULL
LIMIT 100
"""
df_spatial = read_gbq(query_spatial, credentials = credentials)
```

```python
#df_spatial['Latitude'] = float(df_spatial['Latitude'])
df_spatial.info()
```

```python
# Convert the lat/lon to shapely Point objects
df_spatial['geometry'] = df_spatial.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)

# Convert the DataFrame to a GeoDataFrame
points_gdf = gpd.GeoDataFrame(df_spatial, geometry='geometry')
```

```python
# Ensure both GeoDataFrames use the same CRS
points_gdf.set_crs(epsg=4326, inplace=True)
communities.to_crs(epsg=4326, inplace=True)
```
